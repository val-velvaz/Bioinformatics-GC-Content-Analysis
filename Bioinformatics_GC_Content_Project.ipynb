<div style="background-color: #eff0f8; padding: 20px; border-left: 6px solid #f0c5df; border-radius: 8px; margin: 20px 0;">
    <h4 style="margin-top: 0; color: #2c3e50;">💡 Why does GC Content Matter?</h4>
    <p style="margin-bottom: 0; color: #34495e;">In DNA, Guanine (G) and Cytosine (C) are bound by <strong>three hydrogen bonds</strong> (compared to two for Adenine and Thymine). This makes GC-rich regions physically more stable. By calculating how the GC percentage fluctuates across a genome (using a <strong>sliding window algorithm</strong>), bioinformaticians can discover crucial biological features such as <strong>Origins of Replication</strong>, <strong>gene-rich regions</strong>, and <strong>horizontal gene transfer</strong>.</p>
</div>

<div style="background: linear-gradient(90deg, #fce1ec, #f0c5df); padding: 12px 20px; border-radius: 8px; border-left: 6px solid #c95c9e; margin: 15px 0;">
    <h2 style="color: #2c3e50; margin: 0; font-family: 'Segoe UI', sans-serif;">1. Environment setup</h2>
</div>First, we need to ensure that our environment is ready. We will use **Biopython**, the standard library for computational biology in Python.

<div style="background: linear-gradient(90deg, #fce1ec, #f0c5df); padding: 12px 20px; border-radius: 8px; border-left: 6px solid #c95c9e; margin: 15px 0;">
    <h2 style="color: #2c3e50; margin: 0; font-family: 'Segoe UI', sans-serif;">2. Data Loading & Exploration</h2>
</div>
We have downloaded the genomic sequence of the **pBR322 plasmid** from the NCBI Nucleotide database in **FASTA** format. Let's load it into our notebook and inspect its basic properties.

```python
from Bio import SeqIO

# Read the FASTA file
record = SeqIO.read("pBR322.fasta", "fasta")
sequence = record.seq

# Display sequence metadata
print(f"Sequence ID: {record.id}")
print(f"Description: {record.description}")
print(f"Total Length: {len(sequence)} base pairs (bp)")
```

<div style="background-color: #eff0f8; border-left: 6px solid #c95c9e; padding: 20px; border-radius: 8px; text-align: center; margin: 20px 0;">
    <span style="font-size: 1.1em; color: #c95c9e;"><strong>Mathematical Formula:</strong></span>

$$ \color{#2c3e50}{\text{GC Percentage} = \frac{\text{Count of G} + \text{Count of C}}{\text{Total Sequence Length}} \times 100} $$

</div>

```python
# Count the individual nucleotides
num_g = sequence.count("G")
num_c = sequence.count("C")

print(f"Total Guanine (G): {num_g}")
print(f"Total Cytosine (C): {num_c}")

# Calculate the percentage
global_gc_percentage = (num_g + num_c) / len(sequence) * 100
print(f"\nGlobal GC Content: {global_gc_percentage:.2f}%")
```

<div style="background: linear-gradient(90deg, #fce1ec, #f0c5df); padding: 12px 20px; border-radius: 8px; border-left: 6px solid #c95c9e; margin: 15px 0;">
    <h2 style="color: #2c3e50; margin: 0; font-family: 'Segoe UI', sans-serif;">4. The Sliding Window Algorithm</h2>
</div>
A global average doesn't tell us the whole story. To see the genomic architecture, we will slice the DNA into small, overlapping "windows" and calculate the GC content for each specific section.

**Parameters:**
*   `window_size = 500`: We will analyze chunks of 500 base pairs at a time.
*   `step = 100`: The window will "slide" forward by 100 base pairs in each iteration.

```python
window_size = 500
step = 100 

gc_values = []

# Iterate through the DNA sequence
for i in range(0, len(sequence) - window_size, step):
    sample = sequence[i : i + window_size]
    gc_percentage = (sample.count("G") + sample.count("C")) / window_size * 100
    gc_values.append(gc_percentage)

# Display the first 10 window results
print("GC percentages for the first 10 windows:")
print([round(val, 2) for val in gc_values[:10]])
```

<div style="background: linear-gradient(90deg, #fce1ec, #f0c5df); padding: 12px 20px; border-radius: 8px; border-left: 6px solid #c95c9e; margin: 15px 0;">
    <h2 style="color: #2c3e50; margin: 0; font-family: 'Segoe UI', sans-serif;">5. Data Visualization</h2>
</div>
Now we transform our numerical array into a compelling visual story using `matplotlib`. Notice how the GC content drops significantly towards the end of the plasmid, which is a key biological indicator!

```python
import matplotlib.pyplot as plt

# Create a beautiful and professional plot
plt.figure(figsize=(10, 5))
plt.plot(gc_values, color='#34495e', linewidth=2.5, label="Local GC Content")

# Add the baseline (Global Average)
plt.axhline(y=global_gc_percentage, color='#e74c3c', linestyle='--', linewidth=2, label=f'Global Average ({global_gc_percentage:.2f}%)')

# Fill the area under the curve using brand colors for a polished look
plt.fill_between(range(len(gc_values)), gc_values, global_gc_percentage, 
                 where=[v > global_gc_percentage for v in gc_values], 
                 interpolate=True, color='#eae7fa', alpha=0.7)
plt.fill_between(range(len(gc_values)), gc_values, global_gc_percentage, 
                 where=[v <= global_gc_percentage for v in gc_values], 
                 interpolate=True, color='#fce1ec', alpha=0.7)

# Aesthetics and labels
plt.title("GC Content Variation Across pBR322 Plasmid", fontsize=16, fontweight='bold', pad=15)
plt.xlabel(f"Genomic Position (Windows of {window_size} bp)", fontsize=12)
plt.ylabel("GC Percentage (%)", fontsize=12)
plt.legend(loc="upper right", fontsize=10)
plt.grid(True, linestyle=':', alpha=0.7)

# Show the plot
plt.tight_layout()
plt.show()
```
