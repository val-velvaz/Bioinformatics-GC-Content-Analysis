## Assignment 3: Sets — Tumor vs. Healthy Tissue
The Scenario: You ran a pipeline to detect mutated genes. You have a list of mutated genes found in a patient's tumor sample, and another list from a healthy tissue sample. Sequencing is noisy, so your lists have accidental duplicates. You need to find the unique biological signal.

Create: Convert the following lists into sets to automatically remove the duplicates:
* tumor_sample = ["TP53", "EGFR", "PTEN", "TP53", "KRAS"]
* healthy_sample = ["PTEN", "BRCA2", "PTEN"]

```python
tumor_sample = ["TP53", "EGFR", "PTEN", "TP53", "KRAS"]
healthy_sample = ["PTEN", "BRCA2", "PTEN"]

tumor_sample = set(tumor_sample)
healthy_sample = set(healthy_sample)

print(tumor_sample)
print(healthy_sample)
```

Intersection: The lead researcher asks: "Which mutated genes are present in BOTH samples?" Find the intersection.

```python
mutaded_genes = tumor_sample.union(healthy_sample)
print(mutaded_genes)
```

Difference: The oncologist asks: "Which genes are mutated ONLY in the tumor? (This is our target for therapy!)" Find the difference (genes in the tumor set that are NOT in the healthy set).

```python
tumor_only_genes = tumor_sample.difference(healthy_sample)
print(tumor_only_genes)
```
